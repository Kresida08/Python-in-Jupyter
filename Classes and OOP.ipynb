CLASSES AND OBJECT ORIENTED PROGRAMMING

```python
class Program1:

    def __init__ (self,val1,val2):
        self.val1 = val1
        self.val2 = val2

    def method1(self):
        print("This is method 1")
        print ("The values are -" +self.val1 + "  "+self.val2)
```

```python
#Type 1 of inputting the values.
obj = Program1 ("i5" , "16GB")
obj.method1()
#Type 2 of inputting the values.
Program1.method1(obj)
#Type 1.
obj2=Program1("i7" , "8GB")
obj2.method1()
```
