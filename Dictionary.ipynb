DICTIONARY IN PYTHON.
Dictionary in Python is a datastructure with Key-value pairs. Dictionaries are changeable, ordered and do not allow duplicates.

```python
popdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "year" : 1990
}
print(popdict)
```

```python
print (popdict["brand"])  #print's only the mentioned item.
x = popdict.get ("brand") #Give the variable to declare and the item u want to be declared.
print (x)
```

```python
#Will print the key words.
print(popdict.keys())
```

```python
#Single entry to change/update the year.  
popdict["year"]=2023
print (popdict)

#Little complicated entry for the update.
popdict.update({"year":2025})
print (popdict)
```

```python
#Add an element to the list.
popdict.update({"color":"Red"})
print(popdict)
```

```python
#To remove/drop/delete from the list.
popdict.pop("model")
```

```python
#To print elements present in the list.
for a,b in popdict.items():
    print(a,b)
```

```python
#Prints the value in the list.
for x in popdict.values():
    print(x)
```
