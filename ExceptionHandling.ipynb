MULTITHREADING

```python
#Thread is a light-weight process. Helps to achieve mulltitasking.
#CPU has multi-core processor, helps to do multiple tasks in different cores

from time import sleep
from threading import*
class Program1(Thread):
    def run(self):
        for i in range(4):
            print("Hii")
            sleep(1)

class Program2(Thread):
    def run(self):
        for i in range(4):
            print("Hello")
            sleep(1)

t1 = Program1()
t2 = Program2()

#t1.run()
#t2.run()

t1.start()
t2.start() #Starts the thread

t1.join()  #Joins t1 with t2
t2.join()
print("Byee")
```
