LISTS IN PYTHON
-> List is a data structure in python that is mutable, changeable, ordered sequence of elements. Each value in list is called item. List has Square brackets.

```python
#Square bracket shows list.
type ([])
```

```python
#lst is the variable name of the list, list will show list as defined.
lst = list()
type (lst)
```

```python
lst = ["Apple","Mango","Orange","Kiwi"]  #The list is formed.
lst.append("Banana")   #Append->Add at the end of the list.
print(lst)   #Print the above list
```

```python
lst.insert(1,"Watermelon")  #Insert the value in position 1.
print (lst)  #Print the entire o/p after inserting.
print (lst[2])  #Print only the value with indexing 2. Indexing starts from 0,1,2...
```

```python
lst[1: ]  #values from index 1 to last.
```

```python
lst.extend([7,7,7])  #will extend the existing list at the end.
print (lst)  
```

```python
lst.pop()  #Removes top element like a stack.
```

```python
lst.count(7)  #Counts the number of times an element is repeated in the list.
```

```python
lst.count(8)
```
