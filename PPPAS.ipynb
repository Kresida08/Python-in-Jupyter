PUBLIC, PRIVATE AND PROTECTED AND SPECIFIERS

```python
#Define a constructoe with class variables
class Car():
    def __init__ (self, windows, doors, engineType):
        #These are public variables and can be used and initialised anywhere.
        self.windows = windows
        self.doors = doors
        self.engineType = engineType

Audi = Car(4,5,"Diesel")
print (Audi.windows)  #Will print number of windows given.
print (Audi.doors)  #Will print nummber of doors given
print (Audi.engineType)  #Will print the engine type mentioned.
```

```python
#Define a constructoe with class variables
class Book():
    def __init__ (self, pages, cardboard, index, Author):
        #These are protected variables and is defined by "_ ubderscore".
        self._pages = pages
        self._cardboard = cardboard
        self._index = index
        self._Author = Author
#Protected variables are highly used in inheritance.
Manifest = Book(350,2,3, "Roxie Nafousi")
print (Manifest.pages)
print (Manifest.cardboard)
print (Manifest.Author)
print (Manifest.index)
```

```python
#Private variable defined by __ i.e. double underscore. The variable cannot be accessed or modified outside the class.
class Car():
    def __init__ (self, windows, doors, engineType):
        self.__windows = windows
        self.__doors = doors
        self.__engineType = engineType
audi = Car(4,5,"Diesel")
#Print(audi.windows)
print (dir(audi))
```
